```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("netflix_titles.csv")

# Plot: Count of Movies vs TV Shows
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', hue='type', palette='pastel', legend=False)
plt.title("Count of Movies vs TV Shows on Netflix")
plt.xlabel("Type of Content")
plt.ylabel("Count")
plt.show()
```

**Insight:**  
Netflix has more **Movies** than **TV Shows**, highlighting its focus on short-form entertainment content.

```python
from collections import Counter

# Split countries and flatten the list
country_list = df['country'].str.split(', ')
flat_countries = [country for sublist in country_list if isinstance(sublist, list) for country in sublist]

# Count top 10 countries
top_countries = Counter(flat_countries).most_common(10)
country_names = [c[0] for c in top_countries]
country_counts = [c[1] for c in top_countries]

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=country_counts, y=country_names, hue=country_names, palette='Set2', legend=False)
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

**Insight:**  
The **United States**, **India**, and **United Kingdom** are the leading countries producing Netflix content, reflecting both high audience demand and production capacity in these regions.


```python
# Count releases by year
release_counts = df['release_year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(14,6))
sns.lineplot(x=release_counts.index, y=release_counts.values, color='tomato')
plt.title("Netflix Content Release Trend Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles Released")
plt.grid(True)
plt.show()
```

**Insight:**  
Netflix content production **steadily grew**, peaking around **20192020**, reflecting its global expansion and increased content demand during the pandemic era.


```python
# Filter only movies and drop NaN durations
movies_df = df[(df['type'] == 'Movie') & (df['duration'].notnull()) & (df['duration'] != 'Unknown')].copy()

# Remove ' min', handle safely, and convert to int
movies_df['duration_int'] = movies_df['duration'].str.replace(' min', '').astype(int)

# Plot duration distribution
plt.figure(figsize=(10,6))
sns.histplot(data=movies_df, x='duration_int', bins=30, kde=True, color='coral')
plt.title("Distribution of Movie Durations on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

**Insight:**  
The majority of Netflix movies are between **80 to 100 minutes** long, which aligns with standard movie runtime preferences for casual viewers. This also suggests Netflix's focus on delivering content thats easy to consume in a single sitting.


```python
# Filter TV Shows with valid duration
tv_df = df[(df['type'] == 'TV Show') & (df['duration'].notnull()) & (df['duration'] != 'Unknown')].copy()

# Remove text and convert to int
tv_df['seasons'] = tv_df['duration'].str.replace(' Seasons', '')
tv_df['seasons'] = tv_df['seasons'].str.replace(' Season', '')
tv_df['seasons'] = tv_df['seasons'].astype(int)

# Plot season distribution
plt.figure(figsize=(10,6))
sns.countplot(data=tv_df, x='seasons', palette='cool', order=tv_df['seasons'].value_counts().index)
plt.title("Number of Seasons in Netflix TV Shows")
plt.xlabel("Seasons")
plt.ylabel("Number of TV Shows")
plt.show()
```

**Insight:**  
Most TV Shows on Netflix have just **1 season**, suggesting a strong trend toward **limited series** and mini-shows. Only a small percentage of shows are renewed for multiple seasons.
